# In this assignment students have to transform iris data into 3 dimensions
# and plot a 3d chart with transformed dimensions and colour each data
# point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import pandas as pd
```

```python
a = datasets.load_iris()
```

```python
pd.Series(a)
```

```python
df = pd.DataFrame(a['data'],columns=a['feature_names'])
df['target']=a['target']
```

```python
df.head()
```

```python
df1 = df[["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)"]].copy()
```

```python
df1.head()
```

```python
result = np.array(df['target'])
```

```python
tf = decomposition.PCA(n_components=3)
```

```python
transformed = tf.fit_transform(df1)
```

```python
transformed[:,0]
```

```python
fig = plt.figure(num=2,figsize=(10,10))
ax = Axes3D(fig)
p = ax.scatter3D(xs =transformed[:,0] ,ys =transformed[:,1],zs = transformed[:,2],c=result,s=50,cmap=plt.cm.rainbow_r)
ax.set_title("Data Analysis ")
ax.set_xlabel('x',labelpad=20)
ax.set_ylabel('y',labelpad=20)
ax.set_zlabel('z',labelpad=20)
plt.show()
```
